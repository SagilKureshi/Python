- Index of strings

```python
s = "ABCDEF"
print(s[0])
print(s[4])
print(s[2])

```

- Range of String

```python
print(s[1 : 4])
print(s[0 : ])
print(s[ : 5])
print(s[ : ])
print(s[ : :2])

```

- for in loop

```python
for i in s:
    print(i,end=" ")
```

- String Concatination

```python
s = "s " + s[0:]
print(s)
```

Replace String

```python
s1 = s.replace("ABC" , "XYZ")
print(s1)
```

Length of String

```python
print(len(s))
```

- upper() & lower()

```python
print(s.upper())
print(s.lower())
```

- strip()

```python
k = "  abc   "
print(k.strip())
```

-Repetition

```python
print(k*3)
```
